```python
import requests
import pandas as pd
```

```python
API_BASE_URL = "https://archive-api.open-meteo.com/v1/archive"
LATITUDE = 38.5816
LONGITUDE = -121.489906
START_DATE = "2010-09-21"
END_DATE = "2023-09-21"
import requests
import csv
OUTPUT_CSV = "weather_data.csv"

def getWeatherData(latitude, longitude, timeStart, timeEnd, outputCSVPath):
    response = requests.get(API_BASE_URL, params={
        'latitude': latitude,
        'longitude': longitude,
        'start_date': timeStart,
        'end_date': timeEnd,
        'hourly': 'temperature_2m'
    })

    if response.status_code == 200:
        data = response.json()
        times = data['hourly']['time']
        temperatures = data['hourly']['temperature_2m']
        
        # Open the CSV file for writing
        with open(outputCSVPath, 'w', newline='') as csvfile:
            writer = csv.writer(csvfile)
            
            # Write the header row
            writer.writerow(["Time", "Temperature (°C)"])
            
            # Write each row of data
            for time, temperature in zip(times, temperatures):
                writer.writerow([time, temperature])
        
        print(f"Data written to {outputCSVPath}")

    else:
        print(f"Error {response.status_code}: {response.text}")

# Example usage:
getWeatherData(38.5816, -121.489906, "2010-09-21", "2023-09-21")
```
