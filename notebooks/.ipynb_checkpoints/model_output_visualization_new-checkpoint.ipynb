```python
def generate_prediction_figure1(data):
    
    output_length=24
    
    observed = []
    preds = []
    
    plt.figure(figsize=(16, 6))
    
    num_preds = len(data['expected'])//20
    for i in range(0, num_preds):
        source = [sublist[0] for sublist in data['source_readings'][i]]
        pred = data['predicted'][i]
        obs = data['expected'][i]
        
        x_values = range(24*3*(i),24*(3*(i)+1),1)
        plt.plot(x_values, source, marker='.', linestyle='-', c='g')
        
        x_values = range(24*(3*i+1),24*(3*i+1)+output_length,1)
        plt.plot(x_values, pred, marker='.', linestyle='-', c='r')
        plt.plot(x_values, obs, marker='.', linestyle='-', c='b')
        
        observed += obs
#         observed += source
        preds += pred
    
    plt.title('Expected vs. Predicted')
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.show()
```

```python
def rmse(output, target):
    output = np.array(output)
    target = np.array(target)
    loss = np.sqrt(np.mean((output - target) ** 2))
    return loss
```

```python
def generate_month_forecast_figure(path, month_idx, month=0, T=24, save=False):
    plt.figure(figsize=(16, 6))  # Set the figure size for better visibility
    observed = []
    predicted = []
    
    eia_data = pd.read_csv("../data/final_tables/banc2023andcomposite/banc2023andcomposite.csv")
#     mu= -0.003909946350439127
#     sigma= 0.9970714468302634
    data_2023 = eia_data[eia_data["2023"]==1]
#     data_2023["Normalized Forecast"] = (data_2023["Normalized Forecast"] - mu)/sigma

    actual_month = month - 1
#     actual_month = 1+(month+6)%12
    
    eia_forecast_data = data_2023[data_2023["month"]==actual_month]["Normalized Forecast"].tolist()

#     print(data_2023[data_2023["month"]==actual_month].groupby('day').count())
#     t = data_2023[data_2023["month"]==actual_month]
#     day1 = t[t["day"]==2]
#     print(day1)
    
    with open(path, 'r') as file:
        data = json.load(file)

        # Step 1: find start index
        start_idx = 0
        while True:
            source = data['source_readings'][start_idx]
#             for s in source:
#                 print(s)
            m = source[0][month_idx]
            if m == month:
                break
            else:
                start_idx += 1

        idx = start_idx-(1+(T-1)//24)
        print(idx)
        offset = T%24
        end_offset = T%24
        while True:
            source = data['source_readings'][idx+(1+(T-1)//24)]
            expected_ex = data['expected'][idx][offset:]
            predicted_ex = data['predicted'][idx][offset:]
            offset = 0

            m = source[0][month_idx]
            if m != month:
                observed += expected_ex[:end_offset]
                predicted += predicted_ex[:end_offset]
                break
            observed += expected_ex
            predicted += predicted_ex
            
            idx +=1

        plt.plot(observed,linestyle='-', c='r', marker='.', linewidth=1, alpha=0.7, label="Observed")
        plt.plot(predicted,linestyle='-', c='b', marker='.', linewidth=1, alpha=0.7, label="Model 24h Forecast")
        plt.plot(eia_forecast_data,linestyle='-', c='g', marker='.', alpha=0.7, label="EIA 24h Forecast")
#         plt.title(f"C7 2023 Forecast Example for Month {actual_month}")
        plt.xlabel('Hour')
        plt.ylabel('Demand (Normalized)')
        plt.legend()
        if save:
            print("saving")
            print(path)
            code = path.split("/")[2].split("_")[0]
            plt.savefig(f"/Users/aidanwiteck/Desktop/Princeton/Year 4/Thesis/Homeworks/SpringInterimReport/{code}_{actual_month}_predictions.pdf", format="pdf", bbox_inches="tight")
            print("Saved")
        plt.show()
        
        model_rmse = rmse(predicted, observed)
        eia_rmse = rmse(eia_forecast_data, observed)
        
        print(f"Model RMSE: {model_rmse}")
        print(f"EIA RMSE: {eia_rmse}")
```

```python
def generate_month_forecast_figure_168(path, month_idx, month=0, save=False):
    observed = []
    predicted = []
    
    eia_data = pd.read_csv("../data/final_tables/banc2023andcomposite/banc2023andcomposite.csv")
    data_2023 = eia_data[eia_data["2023"]==1]

    actual_month = month+1
    
    eia_forecast_data = data_2023[data_2023["month"]==actual_month]["Normalized Forecast"].tolist()

    
    with open(path, 'r') as file:
        data = json.load(file)

        # Step 1: find start index
        start_idx = 0
        while True:
            source = data['source_readings'][start_idx]
#             for s in source:
#                 print(s)
            m = source[0][month_idx]
            if m == month:
                break
            else:
                start_idx += 1

        idx = start_idx-7
        print(idx)
        while True:
            source = data['source_readings'][idx+7]
            expected_ex = data['expected'][idx]
            predicted_ex = data['predicted'][idx]

            m = source[0][month_idx]
            if m != month:
                break
            observed += expected_ex
            predicted += predicted_ex
            
            idx +=1

        plt.figure(figsize=(16, 6))  # Set the figure size for better visibility
        plt.plot(observed,linestyle='-', c='r', marker='.', linewidth=1, alpha=0.7, label="Observed")
        plt.plot(predicted,linestyle='-', c='b', marker='.', linewidth=1, alpha=0.7, label="Model 24h Forecast")
        plt.plot(eia_forecast_data,linestyle='-', c='g', marker='.', alpha=0.7, label="EIA 24h Forecast")
#         plt.title(f"C7 2023 Forecast Example for Month {actual_month}")
        plt.xlabel('Hour')
        plt.ylabel('Demand (Normalized)')
        plt.legend()
        if save:
            print("saving")
            print(path)
            code = path.split("/")[2].split("_")[0]
            plt.savefig(f"/Users/aidanwiteck/Desktop/Princeton/Year 4/Thesis/Homeworks/SpringInterimReport/{code}_{actual_month}_predictions.pdf", format="pdf", bbox_inches="tight")
            print("Saved")
        plt.show()
        
        model_rmse = rmse(predicted, observed)
        eia_rmse = rmse(eia_forecast_data, observed)
        
        print(f"Model RMSE: {model_rmse}")
        print(f"EIA RMSE: {eia_rmse}")
```
