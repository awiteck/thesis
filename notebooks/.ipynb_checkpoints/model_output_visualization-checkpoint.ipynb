```python
import json
import matplotlib.pyplot as plt
```

```python
# Step 1: Read the data from the JSON file
with open('../model_outputs/2024_2_19_outputtest_example_outputs.json', 'r') as file:
    data = json.load(file)
    
expected_ex = data['expected'][0]
predicted_ex = data['predicted'][0]

# Step 2: Plotting
plt.figure(figsize=(10, 6))  # Set the figure size for better visibility

# Plot "expected" in red
plt.plot(expected_ex, 'r-', label='Expected')

# Plot "predicted" in blue
plt.plot(predicted_ex, 'b-', label='Predicted')

plt.title('Expected vs. Predicted')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Assuming the "expected" and "predicted" data are lists of lists
expected = [item for sublist in data['expected'] for item in sublist]
predicted = [item for sublist in data['predicted'] for item in sublist]

# Step 2: Plotting
plt.figure(figsize=(10, 6))  # Set the figure size for better visibility

# Plot "expected" in red
plt.plot(expected, 'r-', label='Expected')

# Plot "predicted" in blue
plt.plot(predicted, 'b-', label='Predicted')

plt.title('Expected vs. Predicted')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```
