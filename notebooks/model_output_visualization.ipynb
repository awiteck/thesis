```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
idx = 1
# Step 1: Read the data from the JSON file
with open('../model_outputs/2024_2_20_outputtest_example_outputs.json', 'r') as file:
    data = json.load(file)
    
source = [sublist[0] for sublist in data['source_readings'][idx]]
expected_ex = source + data['expected'][idx]
predicted_ex =  source + data['predicted'][idx]

# print(predicted_ex)
# print(expected_ex)

# Step 2: Plotting
plt.figure(figsize=(10, 6))  # Set the figure size for better visibility



# Plot "predicted" in blue
plt.plot(predicted_ex, 'b-', label='Predicted')
# Plot "expected" in red
plt.plot(expected_ex, 'r-', label='Expected')

plt.title('Expected vs. Predicted')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
def generate_prediction_figure1(data):
    
    output_length=24
    
    observed = []
    preds = []
    
    plt.figure(figsize=(16, 6))
    
    num_preds = len(data['expected'])//4
    for i in range(0, num_preds):
        source = [sublist[0] for sublist in data['source_readings'][i]]
        pred = data['predicted'][i]
        obs = data['expected'][i]
        
        x_values = range(24*3*(i),24*(3*(i)+1),1)
        plt.plot(x_values, source, marker='.', linestyle='-', c='g')
        
        x_values = range(24*(3*i+1),24*(3*i+1)+output_length,1)
        plt.plot(x_values, pred, marker='.', linestyle='-', c='r')
        plt.plot(x_values, obs, marker='.', linestyle='-', c='b')
        
        observed += obs
#         observed += source
        preds += pred
    
        
    # Plot "predicted" in blue
#     plt.plot(preds, 'b-', label='Predicted')
    
    # Plot "expected" in red
#     plt.plot(observed, 'r-', label='Expected')

    plt.title('Expected vs. Predicted')
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.show()
```

```python
def generate_prediction_figure2(data):
    '''
    Generate a figure that looks like Fig 13. in L'Heareaux et al's paper. 
    NOTE: DATA must be in chronological order. For example data['source_readings'][0] 
    should be the actual source readings right before data['source_readings'][1]
    Also prediction length must equal source length. Just to make the figure look 
    as expected. 
    '''
    
    observed = []
    preds = []
    
    num_preds = len(data['expected'])
    for i in range(0, num_preds-1):
        source = [sublist[0] for sublist in data['source_readings'][i+1]]
        pred = data['predicted'][i]
        obs = data['expected'][i]
        observed += source
        preds += pred
    plt.figure(figsize=(10, 6))
    
    # Plot "predicted" in blue
    plt.plot(preds, 'b-', label='Predicted')
    
    # Plot "expected" in red
    plt.plot(observed, 'r-', label='Expected')

    plt.title('Expected vs. Predicted')
    plt.xlabel('Index')
    plt.ylabel('Values')
    plt.legend()
    plt.show()
```

```python
generate_prediction_figure(data)
```
